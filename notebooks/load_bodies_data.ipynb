# Prepare bodies data

This notebook fetches all data related to administrative bodies from the OpenParlData API.

```python
# Notebook config
%reset -f
%reload_ext autoreload
%autoreload 2
```

```python
# Imports
import json
import os
from pathlib import Path

import pandas as pd

from util.oparldata_api import get_bodies
```

```python
# Library config
pd.options.mode.copy_on_write = True
```

```python
# Notebook config

# Paths
out_dir = Path('scraped')
bodies_dir = Path(out_dir, 'bodies')

bodies_dir.mkdir(parents=True, exist_ok=True)
```

```python
# Save data
bodies_api_result = get_bodies()

with open(os.path.join(bodies_dir, 'bodies.json'), 'w', encoding='utf-8') as bodies_json_output:
    json.dump(bodies_api_result, bodies_json_output, ensure_ascii=False, indent=2)
```
