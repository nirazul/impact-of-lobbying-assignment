# Prepare interests data

This notebook fetches all data related to persons' interest ties from the OpenParlData API.

```python

# Notebook config
%reset -f
%reload_ext autoreload
%autoreload 2
```

```python
# Imports
import json
import os
import time
from pathlib import Path

import pandas as pd

from util.oparldata_api import get_persons
```

```python
# Library config
pd.options.mode.copy_on_write = True
```

```python
# Notebook config

# Paths
out_dir = Path('scraped')
bodies_dir = Path(out_dir, 'bodies')
interests_dir = Path(out_dir, 'interests')

interests_dir.mkdir(parents=True, exist_ok=True)
```

```python
# Load bodies data
bodies_data = None

with open(os.path.join(bodies_dir, 'bodies.json'), 'r', encoding='utf-8') as bodies_json_output:
    bodies_data = json.load(bodies_json_output)['data']
```

```python
# Save data
for body in bodies_data:
    body_key = body['body_key']
    time.sleep(0.1)
    try:
        json_path = os.path.join(interests_dir, f'interests_{body_key}.json')

        interests_api_result = get_persons(body_key)
        interests_data = interests_api_result['data']
        with open(json_path, 'w', encoding='utf-8') as interests_json_output:
            json.dump(interests_api_result, interests_json_output, ensure_ascii=False, indent=2)

    except Exception as e:
        print(e)
```
