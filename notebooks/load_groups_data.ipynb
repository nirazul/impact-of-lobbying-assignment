# Prepare groups data

This notebook fetches all data related to administrative groups from the OpenParlData API.

```python
# Notebook config
%reset -f
%reload_ext autoreload
%autoreload 2
```

```python
# Imports
import json
import os
from pathlib import Path

import pandas as pd

from util.oparldata_api import get_groups
```

```python
# Library config
pd.options.mode.copy_on_write = True
```

```python
# Notebook config

# Paths
out_dir = Path('scraped')
groups_dir = Path(out_dir, 'groups')

groups_dir.mkdir(parents=True, exist_ok=True)
```

```python
# Save data
groups_api_result = get_groups()

with open(os.path.join(groups_dir, 'groups.json'), 'w', encoding='utf-8') as groups_json_output:
    json.dump(groups_api_result, groups_json_output, ensure_ascii=False, indent=2)
```
