```python
# Imports
import pandas as pd
from util.lang import get_lang
from util.oparldata_api import get_persons
```

```python
response_data = get_persons('CHE')

interests_list = []

for entry in response_data['data']:
    interests_list.append(
        {
            'id': entry['id'],
            'url': get_lang(entry['website_parliament_url']),
            'num_interests': 0 if not entry['interests'] else entry['interests']['meta']['total_records'],
        }
    )

che_df = pd.json_normalize(interests_list)

# Check
che_df
```
